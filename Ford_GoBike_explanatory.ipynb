# Investigation Overview

What are the features affecting trip duration the most (  user type, the Bike type,weekdays)

# Dataset Overview

The data consisted of Duration and attributes of approximately 83530 Trip. The attributes included ride id, rideable type, start station name,start station id,end station name, end station id, member casual, start date, start time, end date, end time, duration, Duration minutes, Duration seconds, start day, end day,day of week.

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
# load in the dataset 
Ford_GoBike=pd.read_csv('ford-gobike-master.csv')
```

## How long do people spend the trip ?


At first I want to know how much the average duration of the trips.
It turned out that the average duration of the trip between 3 and 20 minutes. This means that people fast ride or picnic places nearby

```python
plt.figure(figsize = (10, 7))
plt.style.use('seaborn')
bins = np.arange(0, Ford_GoBike['Duration_minutes'].max()+1, 1)
plt.hist(data = Ford_GoBike, x = 'Duration_minutes', bins = bins);
plt.xlim([0,200])
plt.title('Average Duration ',fontsize = 18);
plt.xlabel('Duration in Minutes');
plt.ylabel('Member Count');
```

```python
plt.figure(figsize = (10, 7))
plt.style.use('seaborn')
bins = np.arange(0, Ford_GoBike['Duration_minutes'].max()+1, 1)
plt.hist(data = Ford_GoBike, x = 'Duration_minutes', bins = bins);
plt.xlim([0,25])
plt.title('Average duration',fontsize = 18);
plt.xlabel('Duration in Minutes');
plt.ylabel('Member Count');
```

## Does the days of the week affect the duration of the trip?

It turns out that weekends are the longest time people spend on the trip .

```python
plt.figure(figsize = (10, 7))
plt.style.use('seaborn')
sb.pointplot(data = Ford_GoBike, x = 'day_of_week', y = 'Duration_minutes', linestyles = "")
plt.ylabel('Duration in minutes');
plt.xlabel('Day of week',fontsize = 18);
plt.title('Duration Per Days of week');
```

### Does the type of bike have an effect on the duration of the trip?

We found that the Type docked bike was the most preferred and people spend longer on it than electronic bike

```python
 count_day_trip=Ford_GoBike.sort_values(by='start_day')
```

```python
plt.figure(figsize = (10, 7))
plt.style.use('seaborn')

sb.pointplot(data = count_day_trip, x = 'start_day', y = 'Duration_minutes', hue = 'rideable_type', palette = 'muted', ci = None);
plt.title('The consume bikes on trips during a month', fontsize = 18);
plt.legend(title = 'bike Type ');
plt.xlabel('Days');
plt.ylabel('Duration in Minutes');
```

 ## Does the type of users have an effect on the duration of the trip?

I want to know whether member users spend more time on the trip than others? And do they prefer a different type of bike than regular users

We found no differences in terms of preference for the type of bike ,But we found big differences in terms of the duration of the trip, because the members users who spend much less time on the trip than others.

```python
plt.style.use('seaborn');
g = sb.FacetGrid(data = Ford_GoBike, col = 'rideable_type', row = 'member_casual',height=4,margin_titles = True);
g.map(plt.scatter, 'start_day','Duration_minutes');
plt.subplots_adjust(top=0.9);
g.fig.suptitle('Influence of Type of users',fontsize = 18);
```
